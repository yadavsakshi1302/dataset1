```python
plt.figure(1, figsize=(15,6))
n=0
for x in ['Two Wheelers', 'Auto Rickshaws' ,'Cars, Taxis, Vans and LMV'  ,'Buses'  ,'Other Non-Motorizd Vehicles(E-rickshaw etc.)'  ,'Total']:
  n += 1
  plt.subplot(1,6,n)
  plt.subplots_adjust(hspace=0.5, wspace=0.5)
  sns.distplot(dataset[x], bins = 20)
  plt.title('Distplot of {}'.format(x))
plt.show()
```
